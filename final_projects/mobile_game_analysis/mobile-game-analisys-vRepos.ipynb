# Анализ рекламных источников мобильной игры

В нашем распоряжении данные о событиях, совершенных в мобильной игре «Космические братья». В ней пользователи строят свою космическую программу и пытаются преуспеть в нелёгком деле колонизации галактики.

Основная монетизация игры — только планируется. Но предполагается, что в приложении будет происходить показ рекламы на экране с выбором типа объекта для постройки.

В датасете представлены данные по игре пользователей на первом уровне. Завершение первого уровня требует от игрока выполнения одного из двух условий:
- Победа над первым врагом
- Реализация проекта - разработка орбитальной сборки спутников

Также предоставлены два датасета с информацией о рекламных активностях.

**Цель исследования**: проанализировать поведение игроков в зависимости от источника перехода, после чего сделать вывод о том, какой из рекламных источников является наиболее перспективным.

**Описание данных**

В датасете содержатся данные первых пользователей приложения — когорты пользователей, которые начали пользоваться приложением в период с 4 по 10 мая включительно.

Основной датасет, который хранится в таблице по адресу `/datasets/game_actions.csv`:
- _event_datetime_ — время события;
- _event_ — одно из трёх событий:
1. building — объект построен,
2. finished_stage_1 — первый уровень завершён,
3. project — проект завершён;

- _building_type_ — один из трёх типов здания:
1. assembly_shop — сборочный цех,
2. spaceport — космопорт,
3. research_center — исследовательский центр;

- _user_id_ — идентификатор пользователя;
- _project_type_ — тип реализованного проекта.

Помимо основного датасета есть два датасета с информацией о рекламных активностях:

1. Датасет, который хранится в таблице по адресу `/datasets/ad_costs.csv`:
- _day_ - день, в который был совершен клик по объявлению;
- _source_ - источник трафика;
- _cost_ - стоимость кликов.

2. Датасет, который хранится в таблице по адресу `/datasets/user_source.csv`:
- _user_id_ - идентификатор пользователя;
- _source_ - источников, с которого пришёл пользователь, установивший приложение.

```python
# пропуски в столбце _building_type_ - либо реализация проекта, либо завершение уровня победой над врагом
# для получения конкретного действия необходимо заменить пропуски в столбце _building_type_ реализацией проекта,
# а оставшиеся пропуски заменить завершением уровня через победу над врагом

game_actions['action'] = (game_actions['building_type']
                         .fillna(game_actions['project_type'])
                         .fillna(game_actions['event']))
```

### Добавление расчетных данных в таблицу `user_source`

Добавим в таблицу `user_source` столбец _first_date_ с датой первого появления пользователя в игре:

```python
user_source = user_source.merge(
    game_actions.sort_values(by=['user_id', 'date']).groupby('user_id')['date'].first(),
on='user_id', how='left')
user_source = user_source.rename(columns={'date':'first_date'})
```

Добавим в таблицу `user_source` столбец _last_date_ с датой последнего посещения пользователем игры:

```python
user_source = user_source.merge(
    game_actions.sort_values(by=['user_id', 'date']).groupby('user_id')['date'].last(),
on='user_id', how='left')
user_source = user_source.rename(columns={'date':'last_date'})
```

Добавим в таблицу `user_source` столбец _days_in_game_ с количеством дней, проведенным в игре:

```python
# при расчете добавляем 1 день, так как пользователь может прийти в игру и покинуть ее в одну и ту же дату
# если не прибавить 1, то время нахождения такого пользователя в игре будет равно 0

user_source['days_in_game'] = (user_source['last_date'] - user_source['first_date']).dt.days + 1
```

Добавим в таблицу `user_source` столбец _date_finished_stage_1_ с датой завершения 1 уровня:

```python
user_source = user_source.merge(
    game_actions[game_actions['event'] == 'finished_stage_1'].sort_values(by=['user_id', 'date'])
    .groupby('user_id')['date'].first(), on='user_id', how='left')
user_source = user_source.rename(columns={'date' : 'date_finished_stage_1'})
```

Добавим в таблицу `user_source` столбец _days_to_finished_stage_1_ с количеством дней, затраченным пользователем на завершение 1 уровня:

```python
# при расчете также добавляем 1 день, так как пользователь мог завершить уровень в день прихода в игру

user_source['days_to_finished_stage_1'] = abs(((user_source['date_finished_stage_1'] - 
                                       user_source['first_date']).dt.days + 1))
```

Повторно выведем первые 10 строк и общую информацию о таблице `user_source` и убедимся, что столбцы были добавлены верно:

```python
user_source.head(10)
```

### Проверка гипотезы о том, что время завершения уровня через реализацию проекта отличается от времени завершения уровня победой над первым врагом

Перед проверкой гипотезы рассмотрим ключевые статистические данные времени, которое потребовалось пользователям для завершения уровня через реализацию проекта и через победу над первым врагом, а также построим диаграмму распределения.

Определим среднее количество, дисперсию и стандартное отклонение времени, которое потребовалось пользователям для завершения уровня **через реализацию проекта**:

```python
print(
    'Среднее время, которое потребовалось пользователям для завершения уровня через реализацию проекта:', 
    user_source[user_source['strategy'].str.contains('satellite_orbital_assembly')]['days_to_finished_stage_1']
    .mean().round()
)
print(
    'Дисперсия времени, которое потребовалось пользователям для завершения уровня через реализацию проекта:', 
    round(
        np.var(user_source[user_source['strategy'].str.contains('satellite_orbital_assembly')]['days_to_finished_stage_1'], 
               ddof=1), 2)
)
print(
    'Стандартное отклонение времени, которое потребовалось пользователям для завершения уровня через реализацию проекта:', 
    round(
        np.std(user_source[user_source['strategy'].str.contains('satellite_orbital_assembly')]['days_to_finished_stage_1'], 
               ddof=1), 2)
)
```

Определим среднее количество, дисперсию и стандартное отклонение времени, которое потребовалось пользователям для завершения уровня **через победу над первым врагом**:

```python
print(
    'Среднее время, которое потребовалось пользователям для завершения уровня через победу над первым врагом:', 
    user_source[~user_source['strategy'].str.contains('satellite_orbital_assembly')]['days_to_finished_stage_1']
    .mean().round()
)
print(
    'Дисперсия времени, которое потребовалось пользователям для завершения уровня через победу над первым врагом:', 
    round(
        np.var(user_source[~user_source['strategy'].str.contains('satellite_orbital_assembly')]['days_to_finished_stage_1'], 
               ddof=1), 2)
)
print(
    'Стандартное отклонение времени, которое потребовалось пользователям для завершения уровня через победу над первым врагом:', 
    round(
        np.std(user_source[~user_source['strategy'].str.contains('satellite_orbital_assembly')]['days_to_finished_stage_1'], 
               ddof=1), 2)
)
```

Построим диаграмму распределения завершения игроками первого уровня:

```python
(
    user_source[user_source['strategy'].str.contains('finished_stage_1')]['last_date']
    .value_counts().plot(figsize=(15,4), color='red')
)
plt.title('Распределение завершения уровня игроками', fontsize=14)
plt.xlabel('Дата')
plt.ylabel('Количество пользователей, шт.')
plt.show()
```

Диаграммы имеют **нормальное распределение**, скошенное вправо, следовательно для проверки гипотезы применим **t-test**.

Сформулируем **нулевую** гипотезу:
- H0 - время завершения уровня через реализацию проекта и через победу над первым игроком не различается;

Соответственно, **альтернативная** гипотеза будет звучать следующим образом:
- H1 - время завершения уровня через реализацию проекта и через победу над первым игроком различается.

Дисперсии отличаются примерно на 36%, поэтому при проведении t-теста применяем параметр **equal_var = False**.

Принимаем уровень статистической значимости в размере **alpha = 5%**.

Для проверки двусторонней гипотезы проведем **t-тест**. Учитывая, что в тесте принимают участие две независимые друг от друга генеральные совокупности, равность которых нужно проверить, для проверки гипотезы необходимо использовать метод _scipy.stats.ttest_ind_:

```python
sample_1 = user_source[user_source['strategy'].str.contains('satellite_orbital_assembly')]['days_to_finished_stage_1']
sample_2 = user_source[~user_source['strategy'].str.contains('satellite_orbital_assembly')]['days_to_finished_stage_1']

# если p-value окажется меньше данного значения - отвергаем нулевую гипотезу
alpha = .05

# применим к выборкам метод dropna() для исключения пропусков при проведении теста,
# зададим параметр equal_var = False, так как дисперсии выборок не равны между собой
results = st.ttest_ind(sample_1.dropna(), sample_2.dropna(), equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу, время завершения уровня через реализацию проекта и через победу над первым игроком различается")
else:
    print("Не получилось отвергнуть нулевую гипотезу, время завершения уровня через реализацию проекта и через победу над первым игроком различается")
```

В ходе t-теста нулевая гипотеза была отвергнута.

Полученный результат подтверждают фактические значения среднего времени на завершение уровня:
- 14 дней потратили пользователи на завершение уровня через реализацию проекта;
- 12 дней потратили пользователи на завершение уровня через победу над первым врагом. 

Время завершения уровня **зависит** от способа завершения.

- Всего за период проведения рекламной компании было затрачено 7604 у.е.;
- 43% всех рекламных затрат произведено 3 мая 2020 года (в первый день рекламной компании);
- Самое большое количество затрат (29.37%) ушло на рекламный источник Yandex.Direct;

### Основные выводы и наблюдения

В ходе анализа было замечено, что:
- Распределение рекламного бюджета достаточно неравномерное - 43% всех рекламных затрат произведено 3 мая 2020 года (в первый день рекламной компании), что является странным;
- Самый популярный рекламный источник - Yandex.Direct, именно через него в игру пришло 4817 (35.48%) пользователей, что вполне ожидаемо, ведь в данный источник вложено 29% всех рекламных затрат;
- Наименее популярный источник - Youtube (19.78% привлеченных пользователей), что также логично, ведь в данный источник было вложено всего 14% рекламных затрат, что вдвое меньше затрат на конкурентов;
- Примечательно, что рекламный источник Youtube является самым дешевым рекламным источником на одного пользователя (CAC = 0.40 у.е.), в то время, как у Facebook показатель CAC почти в 2 раза выше (вероятнее всего, компании стоит пересмотреть распределение рекламного бюджета);
- В среднем пользователи, завершившие игру через реализацию проекта, проводят в игре больше времени;
- Среди пользователей очень популярна постройка зданий (94% событий);
- За время проведения рекламной компании в приложение было привлечено 13576 новых пользователей, что говорит нам о том, что рекламная компания действительно приносит результат.
