Произведем поиск явных дубликатов таблицы и выведем их количество:

Отлично, явные дубликаты отсутствуют. Однако, в столбце _locality_name_, очевидно, присутствуют неявные дубликаты.

Выведем на экран уникальные значения столбца _locality_name_, предварительно изменив тип данных на _str_:

```python
df['locality_name'].astype('str').sort_values().unique()
```

Заметно, например, что поселок может быть через _е_, а может через _ё_. Уберем тип населенногго пункта, оставив только его название, что будет более корректно:

```python
# напишем функцию, которая применит лемматизацию к столбцу 'locality_name' 
# и удалит из него тип населенных пунктов, оставив название
def locality_name_lemmas(df):
    lemmas=df['locality_name']
    extra_words = ['коттеджный поселок ', 'коттеджный посёлок ', 'деревня ',
                   'поселок городского типа ', 'посёлок городского типа',
                   'поселок железнодорожной станции ', 'посёлок железнодорожной станции ',
                   'поселок станции ', 'посёлок станции ',
                   'поселок при железнодорожной станции ', 'посёлок при железнодорожной станции ',
                   'садоводческое некоммерческое товарищество ', 'садовое товарищество ',
                   'городской поселок ', 'городской посёлок ',
                   'поселок ', 'посёлок ', 'село ', ' ', 'имени']

    for word in extra_words:
        if word in lemmas:
            lemmas=lemmas.replace(word, '')
    return lemmas
df['locality_name'] = df.apply(locality_name_lemmas, axis=1)
```

Снова выведем уникальные значения столбца _locality_name_ для проверки результата:
